### Multiple Condition

```python
a =input("Please Enter a mobile brand")
a= a.lower()
if a=='apple' :
    print("Made in USA")
elif a=="samsung" :
    print("Galaxy s26 ultra")
elif a=="nokia" :
    print("very Old phone")
elif a == "vivo" or a== "oppo" or a=="xiaomi" :
    print("Made in china")
elif a== "micromax" :
    print("Made in india")
else :
    print("Not Found")
    
```

```python
x= 'Satyam'
if x.startswith('S') :
    if len(x) == 6 :
        print("The word starts with S and has 6 characters")
    else :
        print("The word starts with S but does not contain 6 characters")
else :
    print("The Word Does not start with S and Does not contain 6 characters")
```

```python
country=input("Name of the country:")
if country==country.upper():
    print("The font entered is ALL CAPITAL")
elif country==country.lower():
    print("The font is all lowercase.")
elif country==country.capitalize():
    print("The font received is appropriate.")
else:
    print("Mixed case received")
print("One of the 4 blocks of code was executed.")
```

```python
x=int(input("Enter a number"))
if(x>10):
    print('i am in if block')
elif(x==6):
# elif will help in checking more than 1 if condition
    print('i am in elif block')
elif(x==0):
    print('Zero number')
else:
    print('i am in else block')
```

### Nested_If

```python
i = 10
if (i >= 10): #  parent if statement

    if (i < 15): # Nested if statement
        print("i is smaller than 15")
    # These expressions are tested only when the parent if returns True.
    if (i < 12):
        print("i is smaller than 12 too")
    else:
        print("i is greater than 15")
```
