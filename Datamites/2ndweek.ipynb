### Nested_If

```python
i = 10
if (i >= 10): #  parent if statement

    if (i < 15): # Nested if statement
        print("i is smaller than 15")
    # These expressions are tested only when the parent if returns True.
    if (i < 12):
        print("i is smaller than 12 too")
    else:
        print("i is greater than 15")
```

```python
#another example of list ---extract the even numbers
mylst =[2,3,4,5,6,7]
mylst1=[]
for i in mylst:
    if i%2==0:
        mylst1.append(i)
print(mylst1)
```

```python
#the above thing using list comprehension
mylst =[2,3,4,5,6,7]
a=[i for i in mylst if i%2==0]
print(a)
```

```python
#eg:- find out which is even and which is odd
mylst =[2,3,4,5,6,7]
mylst2=[]
for i in mylst:
    if i%2==0:
        mylst2.append("Even")
    else:
        mylst2.append("Odd")
print(mylst2)
        
```

```python
#The above same thing using List Comprehension
mylst =[2,3,4,5,6,7]
b=[  "Even" if i%2==0 else "odd" for i in mylst]
print(b)
```

```python
# Index Extraction
def index_extr(word,char):
    mylst=[]
    for i in range(len(word)):
        if word[i]==char:
            mylst.append(i)
    return mylst
```

```python
index_extr('satyam','s')
```

```python
index_extr('satyam','a')
```

### Map Function

```python
# Map Function
#syntax:-
#  map(function_name,iterable)

mylst =[2,3,4,5,6,7]


def even_odd(x):
    if x%2==0:
        return "even"
    else :
        return "odd"
```

```python
even_odd(10)
```

```python
even_odd(7)
```

```python
a=map(even_odd,mylst)
```

```python
list(a)
```

### LAMBDA FUNCTION

```python
mylst =[2,3,4,5,6,7]

b=map(lambda x : "even" if x%2==0 else "odd" , mylst)
list(b)
```
