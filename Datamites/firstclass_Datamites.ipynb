```python
#1---> Auto type casting
print(4 + 5.5 + False)
print(4 + 5.5 + False + True) 
```

```python
#2---> Forced type casting
#integer
print("conversion of int datatype to other")
x= 5
print(float(x))
print(str(x))
print(bool(x))
print("*" * 10)

#float
print("conversion of float datatype to others")
y= 5.5
print(int(y))
print(str(y))
print(bool(y))
print("*" * 10)

#string
print("conversion of string datatype to others")
z= "satyam"
print(bool(z))
mystr='10'
print(type(mystr))
print(int(mystr))
print(float(mystr))
print("*" * 10)

#Booleans
print("Conversion of boolean dtypes into others")
mybool= True
print(int(mybool))
print(float(mybool))
print(str(mybool))
```
