```python
#Integer DataType
import sys
a=13
print(type(a))
print(sys.getsizeof(a))
print(isinstance(a,int))
```

```python
#Float Datatype
import sys
b=13.0
print(type(b))
print(sys.getsizeof(b))
print(isinstance(b,float))
```

```python
#String Datatype
c="satyam"
print(type(c))
print(sys.getsizeof(c))
print(isinstance(c,str))
```

```python
#Boolean Datatype
x=True #-->1
y=False #-->0
print(type(x),type(y))
print(sys.getsizeof(x),sys.getsizeof(y))
print(isinstance(x,bool),isinstance(y,bool))
```
