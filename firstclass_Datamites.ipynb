## Shortcuts

* Run a cell --> shift + enter
* create cell above --> a
* create a cell --> b
* delete a cell --> dd
* undo a deleted cell --> z
* convert a code cell to markdown--> m
* convert a cell to code cell --> y
* copy a cell --> c
* paste a cell --> v
